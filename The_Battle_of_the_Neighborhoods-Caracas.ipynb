### Problem Background


Caracas, is Venezuela's Capital. Here is the main commercial and cultural center of the 
country. Most of gubernamental and private institutions has their main office in this 
city. Caracas has an excellent climate, with an average temperature of 18.6 ° C throughout 
the year, which makes it a destination sought by tourists.


In Caracas there are people of many nationalities and many ethnicities, so you can say 
that this is a multicultural city.The city center is very crowded by commerce, therefore, 
the cost of services and settlement is high. Competitiveness is high, so the creation of a 
new business requires previous studies.


More information visits https://en.wikipedia.org/wiki/Caracas


### Problem Description


A Bar-Restaurant is a commercial establishment where foods, alcoholic and non-alcoholic drinks are sold. 
They are usually aimed at adults, especially young people. The Bar-Restaurant usually have a 
specific style, that together with the structural design give the main characteristics and 
become striking for certain types of customers. 


The problem lies in finding a suitable place to place a new Bar-Restaurant, which is preferred for 
people because of its authenticity and unique style. It must also be located in an 
accessible place, within a busy area and preferably with low competition.


### Requirements of the client:

The client of this project wishes to establish a new bar, aimed at people between 25 and 
50 years old, who are executives and professionals who wish to have fun after leaving the 
office.

The client also wants the bar to be located in an accessible place where you can get 
customers who meet the profile.
